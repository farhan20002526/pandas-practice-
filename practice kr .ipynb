```python
from faker import Faker
import csv

fake = Faker()

# Define the number of records you want in your dataset
num_records = 400

# Define the course names
course_names = ["Graphic Design", "Web Development", "Python Programming", "Data Science", "Video Production"]

# Define the subject names
subject_names = ["Sub1", "Sub2", "Sub3", "Sub4", "Sub5"]

# Open a CSV file to write the data
with open('student_dataset1.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)

    # Write the header row
    header_row = ['Name', 'Class', 'Roll No', 'Course']
    header_row.extend(subject_names)
    writer.writerow(header_row)

    # Generate and write the fake data for the specified number of records
    for _ in range(num_records):
        name = fake.name()
        student_class = fake.random_int(min=1, max=12)
        roll_no = fake.random_int(min=1, max=1000)
        course = fake.random_element(elements=course_names)
        marks = [fake.random_int(min=0, max=100) for _ in range(len(subject_names))]

        # Write the data to the CSV file
        writer.writerow([name, student_class, roll_no, course] + marks)

print(f"Student dataset with {num_records} records created successfully!")
```

# Appling My custom Function on the dataset 

```python
def grade(per):
    Grade="a"
    if per>=80:
        Grade="A"
    elif per>=70:
        Grade="B"
    elif per>=60:
        Grade="C"
    elif per>=50:
        Grade="D"
    elif per>=40:
        Grade="E"
    else:
        Grade="F"
    return Grade
df["Grade"]=df["Percentage"].apply(grade)
```

```python
def Gen(grade):
    gen=None
    if grade == "A" and "B":
        gen="Female"
    else:
        gen="Male"
    return gen
df["Gender"]=df["Grade"].apply(Gen)
```

```python
def grading_system(sub1,sub2,sub3,sub4,sub5):
    total=500
    obt=sub1+sub2+sub3+sub4+sub5
    per = obt/total*100
    if per>=80:
        Grade="A"
    elif per>=70:
        Grade="B"
    elif per>=60:
        Grade="C"
    elif per>=50:
        Grade="D"
    elif per>=40:
        Grade="E"
    else:
        Grade="F"
        
    return total,obt,per,Grade


grading_system(78,88,90,67,89)
```

```python
del df1['obtained Marks']
```

```python
df1[["Total","Obtained Marks","Precentage","Grade"]]=df1[["Sub1","Sub2","Sub3","Sub4","Sub5"]].apply(lambda x:grading_system(*x),axis=1,result_type="expand")
```
